# Filter with More than One Condition (OR)

```python
#Template
df = pd.read_csv("employees.csv", parse_dates =["Start Date", "Last Login Time"]) 
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")

df.head(3)
```

```python
#Return Senior Management True OR Start date less than 19900101
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
#Return Robert And Client Services OR Greater than 20160601
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

#use parantheses to make the condition happend first
df[(mask1 & mask2) | mask3]
```

# The .isin() Method

```python
#Template
df = pd.read_csv("employees.csv", parse_dates =["Start Date", "Last Login Time"]) 
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")

df.head(3)
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
#More effective way to filter multiple values
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

# The .duplicated() Methods

```python
#Template
df = pd.read_csv("employees.csv", parse_dates =["Start Date", "Last Login Time"]) 
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.sort_values("First Name", inplace=True)
df.head(3)
```

```python
# .duplicated() works by "keep"ing the first value unique and make it false(not duplicated) 
# keep = False will return the values that are duplicated
# So we need to negate the result
# ~ = Negation

# return unique values only 
mask = ~df["First Name"].duplicated(keep=False)
df[mask]
```

# The .drop_duplicated() Methods

```python
#Template
df = pd.read_csv("employees.csv", parse_dates =["Start Date", "Last Login Time"]) 
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.sort_values("First Name", inplace=True)
df.head(3)
```

```python
#df.drop_duplicates() will drop if all the collumn and row are exact identical
len(df)
```

```python
len(df.drop_duplicates())
```

```python
df.drop_duplicates(subset=["First Name"], keep="first")
```

```python
#keep = false will only show UNIQUE values (only show once)
df.drop_duplicates(subset=["Team"], keep=False)
```

```python
#drop values if team AND first name are duplicated
df.drop_duplicates(subset=["Team", "First Name"], inplace=True)
```
